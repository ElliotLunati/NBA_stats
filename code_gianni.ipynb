# Régressions Linéaires

## 1) Régression linéaire simple

```python
print(f"Nombre de joueurs à 0 pts : {(df['PTS']==0).sum()}")
print(f"Nombre de joueurs avec un salaire nul : {(df['adjusted_salary']==0).sum()}")
```

On peut donc envisager de passer au logarithme pour des représentation en Log -Niveau par exemple car on ne perd pas trop de données ( seulement 45 via les joueurs n'ayant marqué aucun point)

```python
df_log = df[(df['PTS'] > 0) & (df['adjusted_salary'] > 0)].copy()
df_log['log_PTS'] = np.log(df_log['PTS'])
df_log['log_salary'] = np.log(df_log['adjusted_salary'])

# Regression    Niveau - Niveau
reg_pts = smf.ols('adjusted_salary ~ PTS', data=df_log).fit()
df_log['salary_predicted_pts'] = reg_pts.predict(df_log['PTS'])    # Salaire prédit par le modèle   Niveau - Niveau

# Regression    Niveau - Log
reg_pts_niv_log = smf.ols('adjusted_salary ~ log_PTS', data=df_log).fit()
df_log['salary_predicted_pts_niv_log'] = reg_pts_niv_log.predict(df_log['log_PTS'])  # Salaire prédit par le modèle   Niveau - Log

# Regression    Log - Niveau
reg_pts_log_niv = smf.ols('log_salary ~ PTS', data=df_log).fit()
df_log['log_salary_predicted_pts'] = reg_pts_log_niv.predict(df_log['PTS'])
df_log['salary_predicted_pts_log_niv'] = np.exp(df_log['log_salary_predicted_pts']) # Salaire prédit par le modèle   Log - Niveau

# Regression    Log - Log
reg_pts_log_log = smf.ols('log_salary ~ log_PTS', data=df_log).fit()
df_log['log_salary_predicted_log_pts'] = reg_pts_log_log.predict(df_log['log_PTS'])
df_log['salary_predicted_pts_log_log'] = np.exp(df_log['log_salary_predicted_log_pts'])  # Salaire prédit par le modèle   Log - log
```

```python
# Comparaison du R^2 des différents modèles

tab_R2 = pd.DataFrame({
    'Modèle': ['Niveau–Niveau', 'Niveau–Log', 'Log–Niveau', 'Log–Log'],
    'R2': [ reg_pts.rsquared , reg_pts_niv_log.rsquared , reg_pts_log_niv.rsquared, reg_pts_log_log.rsquared]
})
tab_R2['R2'] = tab_R2['R2'].round(3)
tab_R2_styled = (
    tab_R2.style
    .highlight_max(subset=['R2'], color='lightgreen')
    .format({'R2': '{:.3f}'}) 
)

display(tab_R2_styled)
```

```python
# Visualiser  des différents types de régressions 
plt.figure(figsize=(16, 12))
plt.suptitle(f" Reg. Linéaire Simple : Salaire ~ PTS")

# Regression    Niveau - Niveau
plt.subplot(2, 2, 1)
plt.scatter(df_log['PTS'], df_log['adjusted_salary'], alpha=0.5)
plt.plot(df_log['PTS'], df_log['salary_predicted_pts'],c='red', lw=2)
plt.xlabel('PTS/match')
plt.ylabel('Salaire ($)')
plt.title(f'Prédiction   Niveau - Niveau (R^2 = {reg_pts.rsquared:.3f})')
plt.ticklabel_format(style='plain', axis='both')

# Regression    Niveau - Log
plt.subplot(2, 2, 2)
plt.scatter(df_log['log_PTS'], df_log['adjusted_salary'], alpha=0.5)
plt.plot(df_log['log_PTS'], df_log['salary_predicted_pts_niv_log'],c='red', lw=2)
plt.xlabel('log(PTS/match)')
plt.ylabel('Salaire ($)')
plt.title(f'Prédiction   Niveau - Log (R^2 = { reg_pts_niv_log.rsquared:.3f})')
plt.ticklabel_format(style='plain', axis='both')

# Regression    Log - Niveau
plt.subplot(2, 2, 3)
plt.scatter(df_log['PTS'], df_log['log_salary'], alpha=0.5)
plt.plot(df_log['PTS'], df_log['log_salary_predicted_pts'],c='red', lw=2)
plt.xlabel('PTS/match')
plt.ylabel('log(Salaire) ($)')
plt.title(f'Prédiction   Log - Niveau (R^2 = {reg_pts_log_niv.rsquared:.3f})')
plt.ticklabel_format(style='plain', axis='both')

# Regression    Log - Log
plt.subplot(2, 2, 4)
plt.scatter(df_log['PTS'], df_log['adjusted_salary'], alpha=0.5)
plt.plot(df_log['PTS'], df_log['salary_predicted_pts_log_log'],c='red', lw=2)
plt.xlabel('PTS/match')
plt.ylabel('Salaire ($)')
plt.title(f'Prédiction Log - Log (R^2 = {reg_pts_log_log.rsquared:.3f})')
plt.ticklabel_format(style='plain', axis='both')
```

Que ce soit via les R^2 ou par observasition graphique, on voit que le meilleure modele reste le plus simple : le modèle Niveau - Niveau  
ie           Salaire ~ PTS

```python
# Regression    Niveau - Niveau (Salaire ~ PTS)
reg_pts = smf.ols('adjusted_salary ~ PTS', data=df).fit()
df['salary_predicted_pts'] = reg_pts.predict(df['PTS'])    

# Regression   Niveau - Niveau  (Salaire ~ +/-)
reg_plus_minus = smf.ols('adjusted_salary ~ PLUS_MINUS', data=df).fit()
df['salary_predicted_plus_minus'] = reg_plus_minus.predict(df['PLUS_MINUS'])
```

```python
# Visualiser les prédictions vs valeurs réelles
plt.figure(figsize=(12, 10))

# Regression linéaire simple du Salaire ~ PTS
plt.subplot(2, 2, 1)
plt.scatter(df['PTS'], df['adjusted_salary'], alpha=0.5)
plt.plot(df['PTS'], df['salary_predicted_pts'], c='red', lw=2)
plt.xlabel('PTS/match')
plt.ylabel('Salaire ($)')
plt.title(f'Prédiction   Salaire ~ PTS (R^2 = {reg_pts.rsquared:.3f})')
plt.ticklabel_format(style='plain', axis='both')

# Graphique 2: Distribution des erreurs
plt.subplot(2, 2, 2)
errors = df['adjusted_salary'] - df['salary_predicted_pts']
plt.hist(errors, bins=50, edgecolor='black')
plt.xlabel('Erreur de prédiction ($)')
plt.ylabel('Fréquence')
plt.title('Distribution des erreurs de prédiction')
plt.axvline(x=0, color='r', linestyle='--', linewidth=2)
plt.ticklabel_format(style='plain', axis='x')



# Regression linéaire simple du Salaire ~ (+/-)
plt.subplot(2, 2, 3)
plt.scatter(df['PLUS_MINUS'], df['adjusted_salary'], c='grey', alpha=0.5)
plt.plot(df['PLUS_MINUS'], df['salary_predicted_plus_minus'],c='red', lw=2)
plt.xlabel('PLUS_MINUS / match')
plt.ylabel('Salaire ($)')
plt.title(f'Prédiction   Salaire ~ (+/-) (R^2 = {reg_plus_minus.rsquared:.3f})')
plt.ticklabel_format(style='plain', axis='both')

# Graphique 2: Distribution des erreurs
plt.subplot(2, 2, 4)
errors_plus_minus = df['adjusted_salary'] - df['salary_predicted_plus_minus']
plt.hist(errors_plus_minus, bins=50, color='grey', edgecolor='black')
plt.xlabel('Erreur de prédiction ($)')
plt.ylabel('Fréquence')
plt.title(f'Distribution des erreurs de prédiction (E[erreur]={errors_plus_minus.mean():.3f})')
plt.axvline(x=0, color='r', linestyle='--', linewidth=2)
plt.ticklabel_format(style='plain', axis='x')

plt.tight_layout()
plt.show()
```

Pour estimer le salaire des joueurs, la première composante que l'on pense à étudier peut être le nombre de points inscrits par match par ce joueur. Dans un premier temps, on étudie la régression linéaire simple du salaire sur la moyenne de points inscrits par match, pour obtenir une première prédiction simple du salaire.


Le Plus-Minus est une statistique qui mesure l'impact d'un joueur sur le terrain en comparant les points marqués et encaissés par son équipe lorsqu'il est sur le terrain. Cela permet donc d'observer dans une certaine mesure le poids d'un joueur individuellement, non seulement en attaque (Nombre de points marqués) mais aussi en défense (Nombre de points encaissés), il tient aussi compte de la dimension collective de l'équipe et donc de l'importance du joueur en question dans le bon fonctionnement de l'équipe.
On remarque cependant que la régression linéaire simple avec cette mesure de performance présente de nombreux problème, notamment:  

Un salaire négatif pour certains  
Une dispersion du nuage de point très importante.  
Un R^2 faible, càd que le plus minus explique une faible portion du salaire

De manière générale, les régressions linéaires simples seront insufsisantes pour prédire précisement les salaires.

On a vu que 'PTS' et '+/-' semble avoir un effet (pratique) sur les salaires, mais sont-ils vraiment significatifs statistiquement.  
On va utiliser un test de Student pour le vérifier. 

H0 : " Le nombre de points inscrits en moyenne n'a pas d'effet significatif sur les salaires "   vs.    H1 :" L'effet de 'PTS' sur les salaires est significatif "  

Avec un modèle de régression linéaire simple (Niveau - Niveau) : Salaires = B0 + B1 * PTS + e  
cela revient à tester     H0: " B1=0 "   vs.   H1: "B1 !=0 "  

```python
print(f"Coef de 'PTS' :  {reg_pts.params['PTS']}")         # Significativité pratique de 'PTS' 
print(f"P-valeur de 'PTS' :  {reg_pts.pvalues['PTS']}")    # Significativité statistique de 'PTS'
```

On a p-valeur < 0.01, c'est à dire que l'on rejette l'hypothèse H0 au seuil de 1%, donc l'effet du nombre de points inscrits en moyenne  sur le salaire, est statistiquement significatif au seuil de 1%. 
La p-valeur est même approximativement nulle, donc la variable est significative à un seuil extrèmement faible.

```python
print(f"Coef de '+/-' :  {reg_plus_minus.params['PLUS_MINUS']}")        # Significativité pratique de 'PLUS_MINUS'
print(f"P-valeur de '+/-' :  {reg_plus_minus.pvalues['PLUS_MINUS']}")   # Significativité statistique de 'PLUS_MINUS'
```

On a p-valeur < 0.01, c'est à dire que l'on rejette l'hypothèse H0 au seuil de 1%, donc l'effet du (+/-) sur le salaire, est statistiquement significatif au seuil de 1%.  

Conclusion : Les variables 'PTS' et '+/-' ont un effet statistiquement significatif au seuil de 1% sur 'adjusted_salary' dans les modèles de régression linéaire simple étudiés.  
En pratique, chacune à un effet très fort sur les salaires.   
Une variation de 1 point inscrits en moyenne correspond à une variation de 942 494 USD dans le salaire.   
Une variation de 1 du '+/-' correspond à une variation de 884 067 USD dans le salaire. 

## 2) Régression locale simple (LOWESS)
### Peut on trouver une régression qui permet d'avoir une courbe de tendance plus proche du nuage de points ? 
Plutôt que d’utiliser la méthode des MCO on va utiliser la méthode lowess (locally weighted scatterplot smoothing) qui agrège plusieurs estimations locales tenant compte des valeurs voisines en une estimation globale. 
Cela permet d’obtenir une courbe qui modélisent mieux les valeurs extrêmes et tient plus compte des fluctuations locales des salaires. On peut notamment voir si l'elstacité entre salaire et 'PTS' ou '+/-' n'est pas constante.

```python
from numpy import interp

lowess = sm.nonparametric.lowess
salary_lowess_pts = lowess(df['adjusted_salary'], df['PTS'], frac=0.3)
salary_lowess_plus_minus = lowess(df['adjusted_salary'], df['PLUS_MINUS'], frac=0.3)

PTS_smooth = salary_lowess_pts[:, 0]
salary_smooth_pts = salary_lowess_pts[:, 1]
df['salary_predicted_lowess_pts'] = interp(df['PTS'], PTS_smooth, salary_smooth_pts)


PLUS_MINUS_smooth = salary_lowess_plus_minus[:, 0]
salary_smooth_plus_minus = salary_lowess_plus_minus[:, 1]
df['salary_predicted_lowess_plus_minus'] = interp(df['PLUS_MINUS'], PLUS_MINUS_smooth, salary_smooth_plus_minus)
```

```python
# Visualiser les prédictions vs valeurs réelles
fig = plt.figure(figsize=(16, 12))
fig.suptitle('Prédiction des salaires en fonction des points inscrits selon 2 méthodes d estimations', fontsize=20)
labels = ['MCO','LOWESS']

# Graphique 1: Regression Salaire ~ PTS
plt.subplot(2, 3, 1)
plt.plot(df['PTS'], df['salary_predicted_pts'],c='red', lw=2)
plt.plot(PTS_smooth, salary_smooth_pts, c='lime', lw=2, label="Courbe LOWESS")
plt.scatter(df['PTS'], df['adjusted_salary'], alpha=0.5)
plt.xlabel('PTS/match')
plt.ylabel('Salaire ($)')
plt.axvline(x=np.median(df['PTS']), color='black', linestyle='--', linewidth=2)
plt.axvline(x=np.percentile(df['PTS'], 95), color='black', linestyle='--', linewidth=2)  # Centile 95 : 95 % des joueurs inscrivent moins de pts que ce seuil
plt.title(f'Régression Salaire ~ PTS')
plt.ticklabel_format(style='plain', axis='both')

# Graphique 2: Distribution des erreurs (MCO)
plt.subplot(2, 3, 2)
errors = df['adjusted_salary'] - df['salary_predicted_pts']
plt.hist(errors, bins=50, edgecolor='black')
plt.xlabel('Erreur de prédiction ($)')
plt.ylabel('Fréquence')
plt.title('Distribution des erreurs de prédiction (MCO)')
plt.axvline(x=0, color='r', linestyle='--', linewidth=2)
plt.ticklabel_format(style='plain', axis='x')

# Graphique 3: Distribution des erreurs (Lowess)
plt.subplot(2, 3, 3)
errors = df['adjusted_salary'] - df['salary_predicted_lowess_pts']
plt.hist(errors, bins=50, edgecolor='black')
plt.xlabel('Erreur de prédiction ($)')
plt.ylabel('Fréquence')
plt.title('Distribution des erreurs de prédiction (Lowess)')
plt.axvline(x=0, color='lime', linestyle='--', linewidth=2)
plt.ticklabel_format(style='plain', axis='x')



# Graphique 4: Regression Salaire ~ PTS
plt.subplot(2, 3, 4)
plt.plot(df['PLUS_MINUS'], df['salary_predicted_plus_minus'],c='red', lw=2)
plt.plot(PLUS_MINUS_smooth, salary_smooth_plus_minus, c='lime', lw=2, label="Courbe LOWESS")
plt.scatter(df['PLUS_MINUS'], df['adjusted_salary'], c='grey', alpha=0.5)
plt.xlabel('(+/-)/Match')
plt.ylabel('Salaire ($)')
plt.title(f'Régression Salaire ~ (+/-)')
plt.ticklabel_format(style='plain', axis='both')

# Graphique 5: Distribution des erreurs (MCO)
plt.subplot(2, 3, 5)
errors = df['adjusted_salary'] - df['salary_predicted_plus_minus']
plt.hist(errors, bins=50, color='grey', edgecolor='black')
plt.xlabel('Erreur de prédiction ($)')
plt.ylabel('Fréquence')
plt.title('Distribution des erreurs de prédiction (MCO)')
plt.axvline(x=0, color='r', linestyle='--', linewidth=2)
plt.ticklabel_format(style='plain', axis='x')

# Graphique 6: Distribution des erreurs (Lowess)
plt.subplot(2, 3, 6)
errors = df['adjusted_salary'] - df['salary_predicted_lowess_plus_minus']
plt.hist(errors, bins=50, color='grey', edgecolor='black')
plt.xlabel('Erreur de prédiction ($)')
plt.ylabel('Fréquence')
plt.title('Distribution des erreurs de prédiction (Lowess)')
plt.axvline(x=0, color='lime', linestyle='--', linewidth=2)
plt.ticklabel_format(style='plain', axis='x')

fig.legend(labels, loc='center', bbox_to_anchor=(0.5, 0.94), ncol=4)
plt.show()
```

On observe bien que l'élasticité du salaire par rapport aux 'PTS' / '+/-' n'est pas constante dans les 2 modèles d'estimation avec LOWESS. On observe que le salaire prédit par le modèle LOWESS, est presque constant pour les joueurs ayant un '+/-' négatif ou nul, et que le salaire explose pour les joueurs pour lequels il est fortement positif. En pratique les joueurs avec un '+/-' positif sont les bons ou très bon joueurs qui sont donc grassement payés, si les joueurs ont un '+/-' négatif, cela signifie qu'il font en moyenne 'perdre' l'équipe et leur valeur diminue en conséquence.  
  
Pour le modèle Salaire ~ PTS, avec LOWESS, on remarque que la courbe de tendance est convexe, l'effet marginal d'un point supplémentaire sur le salaire est beaucoup plus élevé lorsque l'on inscrit en moyenne plus de 20 points par match que lorsque l'on en inscrit moins de 5. Il est en effet beaucoup plus difficile d'augmenter son apport offensif lorsqu'il est déjà élevé car le temps de jeu est limité à 48 minutes par match notamment, donc seuls les meilleurs sont capables d'inscrire plus de 20 points par match ce qui se reflètent dans leur salaire.  
On peut par exemple définir arbitrairement 3 catégories de joueurs : les "role players", les bons joueurs et les "franchise player".  
Les "roles players" sont souvent des joueurs remplacant qui ne possèdent que peu de temps de jeu et qui ne peuvent généralement pas inscrire une forte quantité de points. Dans notre cas, les bons joueurs sont des joueurs titulaires ou emplacants avec un temps de jeu conséquent et qui ont donc la possibilité d'inscrire plus de points que les "role player" mais pas autant que les "franchise players", ils sont également mieux payés que les "role player". Les "franchise players" quant à eux sont des joueurs stars, ce sont généralement les leaders offensifs de leur équipe, ce qui leur permet d'inscire un nombre conséquent de paniers et donc de points. On compte généralement un "franchise player" par équipe (pour simplifier on considère que le top 5% des meilleurs scoreurs sont les "franchise player"), c'est aussi généralement sur lui que l'équipe investit le plus, ces joueurs ont la plupart du temps les salaires les plus élevés.   
  
Sur le graphique 1, avant la première droite vertivale (Mediane des points inscrits) on lit que les salaires prédits des "role player" sont plutôt constant, on observe une augmentation des rémunérations pour les bons joueurs et enfin une explosion des salaires pour les meilleurs scoreurs, ceux qui inscrivent plus de 21 pts par match (plus que 95% des joueurs de la ligue).  

## 3) Régressions linéaires multiples
### Les statistiques individuelles habituelles de jeu (PTS, REB, AST, STL, BLK) prédisent-elles les salaires des joueurs? 
On va désormais considérer une régression linéaire multiple qui tient compte des 5 principales statistiques ( PTS, REB, AST, STL, BLK ):  
          Salaire = B0 + B1xPTS + B2xREB + B3xAST + B4xSTL + B5xBLK + e

```python
reg5 = smf.ols('adjusted_salary ~ PTS + REB + AST + STL + BLK', data=df).fit()
df['salary_predicted_5'] = reg5.predict(df[['PTS','REB','AST','STL','BLK']])

print(f"Erreur moyenne = {(reg5.resid.mean()).round(5)}") # On a un estimateur sans biais 
```

On souhaite savoir si les statistiques individuelles habituelles ont un impact sur le salaire, c'est à dire si au moins l'une d'elle permet d'expliquer le salaire, lorqu'on intègre les 5 au modèle.  
On cherche donc à faire un test de Fisher: 

H0 : " B1 = B2 = B3 = B4 = B5 = 0 "   vs    H1 : " (B1 != 0) ou (B2 != 0) ou (B3 != 0) ou (B4 != 0) ou (B5 != 0) "


```python
reg5.summary()
```

On observe que toutes les p-valeurs sont approximativement nulles, ce qui signifie que l'on rejette H0, donc ces 5 indicateurs ont un effet conjointement significatif sur le salaire.  
Ces 5 statistiques permettent donc de prédire en partie le salaire des joueurs.  
Cependant on observe que le nombre moyen d'interceptions par match est lié négativement au salaire dans notre modèle, ce qui semble incohérent. Une interception n'est pas une action négative au basket-ball, et a même un effet positif pour la victoire de son équipe, il est donc étonnant qu'elle impacte négativement le salaire des joueurs. 

```python
# Visualiser les prédictions vs valeurs réelles
plt.figure(figsize=(16, 6))
plt.suptitle(f"RLM :  Salaire ~ PTS + REB + AST + STL + BLK ( R^2 = {reg5.rsquared:.2f}) ")

# Graphique testant la qualité de la RLM
plt.subplot(1, 2, 1)
plt.scatter(df['adjusted_salary'], df['salary_predicted_5'], alpha=0.5)
plt.plot([df['adjusted_salary'].min(), df['adjusted_salary'].max()],
         [df['adjusted_salary'].min(), df['adjusted_salary'].max()], 'r--', lw=2)
plt.xlabel('Salaire réel ($)')
plt.ylabel('Salaire prédit ($)')
plt.title('Prédictions vs Valeurs réelles ')
plt.ticklabel_format(style='plain', axis='both')

# Graphique 2: Distribution des erreurs
plt.subplot(1, 2, 2)
errors_5 =df['adjusted_salary'] - df['salary_predicted_5']
plt.hist(errors, bins=50, edgecolor='black')
plt.xlabel('Erreur de prédiction ($)')
plt.ylabel('Fréquence')
plt.title(f'Distribution des erreurs de prédiction (E[erreur]={errors_5.mean():.3f})')   # On a un estimateur sans biais
plt.axvline(x=0, color='r', linestyle='--', linewidth=2)
plt.ticklabel_format(style='plain', axis='x')
```

On observe que l'on a beaucoup de mal à prédire les salaires élevés, on ne prédit aucun salaire > 35 Millions , pourtant les salaires réels augmentent quant à eux jusqu'à presque 60 Millions. En prédisant les classements des salaires en fonction des performances en valeurs, on obtient des rangs prédits négatifs, ce qui est impossible.
