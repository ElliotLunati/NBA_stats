```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Charger la base de données fusionnée
df = pd.read_csv('./data/merged_data.csv')

print(f"Dataset chargé: {len(df)} lignes, {len(df.columns)} colonnes")
print(f"\nAperçu des colonnes:")
print(df.columns.tolist())
```

```python
# Afficher les premières lignes
print("\n Aperçu des données:")
print(df.head())

print("\n Informations sur les données:")
print(df.info())
```

```python
adj_salary_season = df.groupby('Season')['adjusted_salary']   # Salaire ajusté par saison 
display(adj_salary_season.describe().round(2))
```

```python
plt.plot(adj_salary_season.max(), c='red',linewidth = 1, label='Max')
plt.plot(adj_salary_season.mean(), c='blue',linewidth = 1, label='Mean')
plt.plot(adj_salary_season.median(), c='green',linewidth = 1, label='Median')
plt.plot(adj_salary_season.min(), c='yellow',linewidth = 1, label='Min')
plt.title("NBA salary trends (2000-2025)")
plt.xlabel("Season")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
avg_points_season = df.groupby('Season')['PTS']
avg_rebounds_season = df.groupby('Season')['REB']
avg_assists_season = df.groupby('Season')['AST']
avg_steals_season = df.groupby('Season')['STL']
avg_blocks_season = df.groupby('Season')['BLK']

df['total_minutes'] =  df['MIN'] * df['GP']
time_played_season = df.groupby('Season')['total_minutes']    # Nombre de minutes jouées par saison en moyenne
```

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Statistical trends by season (2020-2025)', fontsize=20)
labels = []

axes[0, 0].set_title('Points/game evolution') 
axes[0, 0].plot(df.groupby('Season')['PTS'].max(), c='red',linewidth = 1, label='Max')
axes[0, 0].plot(df.groupby('Season')['PTS'].mean(), c='blue',linewidth = 1, label='Mean')
axes[0, 0].plot(df.groupby('Season')['PTS'].median(), c='green',linewidth = 1, label='Median')
axes[0, 0].plot(df.groupby('Season')['PTS'].min(), c='yellow',linewidth = 1, label='Min')
labels.extend(['Max', 'Mean', 'Median', 'Min'])

axes[0, 1].set_title('Rebounds/game evolution')
axes[0, 1].plot(df.groupby('Season')['REB'].max(), c='red',linewidth = 1, label='Max')
axes[0, 1].plot(df.groupby('Season')['REB'].mean(), c='blue',linewidth = 1, label='Mean')
axes[0, 1].plot(df.groupby('Season')['REB'].median(), c='green',linewidth = 1, label='Median')
axes[0, 1].plot(df.groupby('Season')['REB'].min(), c='yellow',linewidth = 1, label='Min')

axes[0, 2].set_title('Assists/game evolution')
axes[0, 2].plot(df.groupby('Season')['AST'].max(), c='red',linewidth = 1, label='Max')
axes[0, 2].plot(df.groupby('Season')['AST'].mean(), c='blue',linewidth = 1, label='Mean')
axes[0, 2].plot(df.groupby('Season')['AST'].median(), c='green',linewidth = 1, label='Median')
axes[0, 2].plot(df.groupby('Season')['AST'].min(), c='yellow',linewidth = 1, label='Min')

axes[1, 0].set_title('Steals/game evolution')
axes[1, 0].plot(df.groupby('Season')['STL'].max(), c='red',linewidth = 1, label='Max')
axes[1, 0].plot(df.groupby('Season')['STL'].mean(), c='blue',linewidth = 1, label='Mean')
axes[1, 0].plot(df.groupby('Season')['STL'].median(), c='green',linewidth = 1, label='Median')
axes[1, 0].plot(df.groupby('Season')['STL'].min(), c='yellow',linewidth = 1, label='Min')

axes[1, 1].set_title('Blocks/game evolution')
axes[1, 1].plot(df.groupby('Season')['BLK'].max(), c='red',linewidth = 1, label='Max')
axes[1, 1].plot(df.groupby('Season')['BLK'].mean(), c='blue',linewidth = 1, label='Mean')
axes[1, 1].plot(df.groupby('Season')['BLK'].median(), c='green',linewidth = 1, label='Median')
axes[1, 1].plot(df.groupby('Season')['BLK'].min(), c='yellow',linewidth = 1, label='Min')

axes[1, 2].set_title('Time played evolution')
axes[1, 2].plot(time_played_season.max(), c='red',linewidth = 1, label='Max')
axes[1, 2].plot(time_played_season.mean(), c='blue',linewidth = 1, label='Mean')
axes[1, 2].plot(time_played_season.median(), c='green',linewidth = 1, label='Median')
axes[1, 2].plot(time_played_season.min(), c='yellow',linewidth = 1, label='Min')


fig.legend(labels, loc='upper center', bbox_to_anchor=(0.5, 0.94), ncol=4)
plt.show()
```
